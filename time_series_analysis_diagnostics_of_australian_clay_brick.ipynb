```r
### 3

brick.dat <- read.table('brick.txt')

# a)
library(TSA)
library(forecast)
# Time series data
brick.ts <- ts(brick.dat, start = 1956, frequency = 4)

# Time series plot
plot(brick.ts, ylab = ' Number of bricks (in millions)', xlab = 'Year', type='o', main=' Time series plot of 
     Australian clay brick production')

# Trend with nonconstant variance
# seasonal pattern, not very clear though

# Decompose the time series data to clearly see its components
brick.dc <- decompose(brick.ts)
plot(brick.dc) 
# Now, it's clear the series has trend+seasonal+random components
```

```r
# b)

# Box-Cox transformation

br.bc <- BoxCox.ar(brick.ts)
br.bc$mle
#[1] 0.4 - max. likelihood estimate of power transformation parameter value, lambda. 
# Then, we choose lambda= 0.5 : Square root transformation

brick.sqrt <- sqrt(brick.ts)
```

```r
# c)

par(mfrow=c(2,2))

plot(brick.sqrt, ylab = 'Number of bricks (sqrt transformation)', xlab = 'Year', type='o', main=' Time series plot of 
     Australian clay brick production') # Still has trend and seasonal pattern

plot(diff(brick.sqrt), ylab = 'First differences', xlab = 'Year', type='o', main=' Time series plot of 
     Australian clay brick production') # Seasonal pattern still quite pronounced

plot(diff(brick.sqrt, lag = 4), ylab = 'First seasonal differences (s=4)', xlab = 'Year', type='o', main=' Time series plot of 
     Australian clay brick production') # Doesn't look stationary

plot(diff(diff(brick.sqrt), lag = 4), ylab = 'Combined first differences', xlab = 'Year', type='o', main=' Time series plot of 
     Australian clay brick production') # Now, looks quite stationary

# We take seasonal and simple differences
brick.diff <- diff(diff(brick.sqrt), lag = 4)

# acf and pacf plots
acf(as.vector(brick.diff), main='Sample ACF', lag.max = 32)
# seasonal lags: spikes at lag 4, 8, 25(strange), 32
# simple lags: spike at lag 1 or no spike (it's close to lag 0)

pacf(as.vector(brick.diff), main="Sample PACF", lag.max = 32)
# seasonal lags: spikes at lag 4, 8, 12, 16, 28
# simple lags: spike at lag 1 or no spike (it's close to lag 0)

# Possible candidates: SARIMA(0,1,0)x(0,1,2)? SARIMA(0,1,0)X(4,1,0)?
```

```r
# d)

# Automatic model selection
auto.arima(brick.sqrt, ic='aic') # ARIMA(2,0,1)(0,1,1)[4]

auto.arima(brick.sqrt, ic='bic') # ARIMA(2,0,0)(0,1,1)[4] 

# Compare all candidate models using aic as criteria
arima(brick.sqrt,order=c(0,1,0),method='ML',seasonal=list(order=c(4,1,0),period=4)) # log likelihood = -122.47,  aic = 252.94
# Potential choice

arima(brick.sqrt,order=c(0,1,0),method='ML',seasonal=list(order=c(0,1,2),period=4)) # log likelihood = -123.25,  aic = 250.49
# Potential choice

arima(brick.sqrt,order=c(2,0,1),method='ML',seasonal=list(order=c(0,1,1),period=4)) # log likelihood = -120.79,  aic = 249.57
# Not taking this because it doesn't consider the simple differencing

arima(brick.sqrt,order=c(2,0,0),method='ML',seasonal=list(order=c(0,1,1),period=4)) # log likelihood = -121.26,  aic = 248.51
# Not taking this because it doesn't consider the simple differencing

brick.ml.010x410 <- arima(brick.sqrt,order=c(0,1,0),method='ML',seasonal=list(order=c(4,1,0),period=4)) 
# This is my choice model. The reasons are given below

brick.ml.010x410


# Call:
#   arima(x = brick.sqrt, order = c(0, 1, 0), seasonal = list(order = c(4, 1, 0), 
#                                                             period = 4), method = "ML")
# 
# Coefficients:
#         sar1     sar2     sar3     sar4
#       -0.8249  -0.8390  -0.5330  -0.3290
# s.e.   0.0780   0.0935   0.0936   0.0772
# 
# sigma^2 estimated as 0.2889:  log likelihood = -122.47,  aic = 252.94

brick.ml.010x012 <- arima(brick.sqrt,order=c(0,1,0),method='ML',seasonal=list(order=c(0,1,2),period=4))
brick.ml.010x012

# Let's do a quick residual check for the 2 potential choices

checkresiduals(brick.ml.010x012) # p-value = 0.001512 for the Ljung-Box test - doesn't behave like a white noise

checkresiduals(brick.ml.010x410) 
# p-value = 0.1062 for the Ljung-Box test - this shows that the residuals behave like a white noise
# Better ACF plot for the residuals
# Better histogram plot and residuals plot looks more stationary

# Mathematical (equation) expression of the fitted model

# (1+0.8249B^4 + 0.8390B^8 + 0.5330B^12 + 0.3290B^16)(1-B)(1-B^4)sqrt(Xt) = Zt
```

```r
# e)

# Diagnostics for the fit of the chosen model 

# Check residuals for the selected model

checkresiduals(brick.ml.010x410) # Quick check of the residuals p-value = 0.1062

# Sample ACF and histogram of the residuals quite good. The ACF plot has 2 spikes at lag 5 & lag 12
# The residuals plot looks quite stationary
# The p-value supports the stationarity of the series

# Standardised residuals plot to check the residuals
plot(rstandard(brick.ml.010x410), xlab="Year", ylab="Standardised residuals", type='o', main = 'Standardised residuals plot')
abline(h=0)
# The plot looks stationary with mean zero

# Normality check for the residuals
hist(rstandard(brick.ml.010x410),xlab="Standardised residuals",main="Histogram plot of the residuals")
#Looks fairly okay

qqnorm(rstandard(brick.ml.010x410),main="Q-Q plot of the residuals")
qqline(rstandard(brick.ml.010x410))
# Doesn't look good. Several points out of the fitting line

# Diagnostic Plots for Time-Series Fits
tsdiag(brick.ml.010x410, gof=20, omit.initial=F)
# The standardized residuals, ACF of residuals and p-values for Ljung-Box statistic all look good
# The residual output fully supports SARIMA(0,1,0)X(4,1,0)[4] model 

Box.test(residuals(brick.ml.010x410),lag =10, type="Ljung-Box",fitdf=2)  # Box-Ljung test  with fitdf=p+q
# p-value for the Ljung-Box does not reject the null hypothesis. It confirms that the residuals behave like a white noise

# Box-Ljung test
# 
# data:  residuals(brick.ml.010x410)
# X-squared = 9.0125, df = 8, p-value = 0.3412
```
