```r
### Time Series Analysis
### Lab 1

### 6
library(TSA)

data("co2")
head(co2)

### a)

# Time series plot for co2 data
plot(co2, main = "Plot of CO2 Levels over Time", xlab="Year", ylab="CO2 levels", type="l")
points(y=co2, x=time(co2), pch=as.vector(season(co2)), cex=0.75)

```

```r
### b)

# Regression model for the trend

model.co2 <- lm(co2 ~ time(co2))
summary(model.co2)

```

```r
### c)

plot(co2, main = "Plot of CO2 Levels over Time", xlab="Year", ylab="CO2 levels", type="l")
points(y=co2, x=time(co2), pch=as.vector(season(co2)), cex=0.75)
abline(model.co2) # least squares fit
```

```r
### d)

plot(y=resid(model.co2), x=as.vector(time(co2)), main = "Plot of residauls vs time", xlab="Year", ylab="residuals", type="l")
points(y=resid(model.co2), x=as.vector(time(co2)), pch=as.vector(season(co2)), cex=0.75)
```

```r
### e)

# sample autocorrelations plot of the residuals
acf(resid(model.co2))
```
